```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt


iris = load_iris()
X = iris.data


kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)


distances = np.linalg.norm(X - kmeans.cluster_centers_[kmeans.labels_], axis=1)


threshold = np.mean(distances) + 2 * np.std(distances)

anomalies = np.where(distances > threshold)[0]

plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, cmap='viridis')
plt.scatter(X[anomalies, 0], X[anomalies, 1], color='red', label='Anomalies', marker='X')
plt.title('K-means Clustering with Anomalies')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.legend()
plt.show()


anomalies
```
