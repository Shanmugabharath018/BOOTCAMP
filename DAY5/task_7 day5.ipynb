```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
iris = load_iris()
X = iris.data 

init_methods = ['random', 'k-means++']

plt.figure(figsize=(12, 6))

for i, init_method in enumerate(init_methods, 1):
   
    kmeans = KMeans(n_clusters=3, init=init_method, n_init=10, random_state=42)
    kmeans.fit(X)
    
    
    pca = PCA(n_components=2)
    X_pca = pca.fit_transform(X)
    
 
    plt.subplot(1, 2, i)
    plt.scatter(X_pca[:, 0], X_pca[:, 1], c=kmeans.labels_, cmap='viridis', label="Clusters")
    plt.title(f'KMeans with {init_method.capitalize()} Initialization')
    plt.xlabel('PCA Component 1')
    plt.ylabel('PCA Component 2')

plt.tight_layout()
plt.show()
```
